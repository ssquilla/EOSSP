```python
def getSamples(sample_path,name=None,instance=None,Ncoeur=None,algo=None,bestCoeur=False,sat=None,seed_instance=None):
    samples = []
    for root,dirs,files in os.walk(sample_path):
        sorted_files = sorted(files)
        #print(sorted_files)
        for file in sorted_files:
            try:
                with open(sample_path+'/'+file,'rb') as handle:
                    dico_hist  = pickle.load(handle)
                    if filtreUnitaire(dico_hist,name=name,instance=instance,Ncoeur=Ncoeur,algo=algo,sat=sat,seed_instance=seed_instance):
                        dico_hist['file'] = str(file)
                        samples.append(dico_hist)
            except Exception as e:
                print(e)
    return samples

def getFirstIteration(sample):
    hist = sample['hist']
    flag = 1
    try:
        firstIt = [h for h in hist if h[1]==flag][0]
    except:
        firstIt = hist[-1]
    duree,obj = firstIt[0],firstIt[2]
    return duree,obj

def cleHistorique(historique):
    return (historique.instance,historique.algo_name,int(historique.Ncoeur))
#samples = sorted(samples,key=cleHistorique)

def getCCA(dico):
    if dico['stats']['#CCA'] is not None:
        return dico['stats']['#CCA']
    else:
        return 0

def reduire(sample,tlim):
    s = deepcopy(sample)
    if s['algo']!="globalSolver":
        s['hist'] = [h for h in s['hist'] if h[0]<=tlim]
        m = max(s['hist'],key=lambda h:h[2])
        s['modes'] = m[-1]
        s['obj'] = m[2]
    return s

#call operateurs, call perturb
def compteursIterations(sample):
    compteur_iterations = {}
    compteur_operateurs = {}
    best_coeur = max([x for x in sample['hist']],key = lambda x : x[2][0])[5]
    for x in sample['hist']:
        if x[5]==best_coeur:
            if x[5] not in compteur_iterations:
                compteur_iterations[x[5]] = 0
                compteur_operateurs[x[5]] = 0
            if x[1]=='END_OPERATEUR' and x[0]>0:
                compteur_operateurs[x[5]] += 1
            if x[1]=="END_ITERATION" and x[0]>0:
                compteur_iterations[x[5]] +=1
    call_ops = sum([compteur_operateurs[coeur] for coeur in compteur_operateurs])
    call_its = sum([compteur_iterations[coeur] for coeur in compteur_iterations])
    if sample['algo']=='LNS':
        call_ops -= call_its
    return call_ops,call_its

def courbesObjectif(hist,tlim):
    return [x[0] for x in hist['hist'] if x[0]<=tlim],[x[2][0] for x in hist['hist'] if x[0]<=tlim]

def getObj(dico_hist):
    return max([h[2] for h in dico_hist['hist']])
    return dico_hist['obj']

def cmpInstance(i1,i2):
    return i1.split('.pb')[0].split('_')[:8] == i2.split('.pb')[0].split('_')[:8]

def cmpRepartition(instance_name,rep):
    return getRepartitionRequete(instance_name)==rep
    
def getRatio(dico_hist):
    return dico_hist['stats']['ratio']

def formatInstance(instance):
    res = instance.split("_")
    return res[3]+"\_"+res[4]+"\_"+res[5]+"\_"+res[6]

def getDateFin(dico_hist):
    return dico_hist['hist'][-1][0]

def getNActivites(dico):
    return dico['stats']['#A']

def profondeurModeMoyen(dico):
    return np.mean([x[1] for x in dico['modes']])

def profondeurModeMax(dico):
    try:
        return max([x[1] for x in dico['modes']])
    except:
        return np.Inf

def getInstance(dico_hist):
    return dico_hist['instance']

def getAlgo(dico_hist):
    if dico_hist['algo']!='main':
        return dico_hist['algo']
    else:
        return extractConfig(dico_hist," solver ",str)

def getCoeurs(dico_hist):
    return dico_hist['coeurs']

def getSatellites(dico_hist):
    return getSatFromName(dico_hist['instance'])

def getSatFromName(name):
    sat = name.split("_")[1]
    tmp = name.split("x")
    s1 = int(tmp[0].split("_")[-1])
    s2 = int(tmp[1].split("_")[0])
    return s1*s2

def filtreUnitaire(hist,name=None,instance=None,Ncoeur=None,algo=None,sat=None,seed_instance=None):
    test_name = (name is None or cmpInstance(getInstance(hist),name))
    test_instance = (instance is None or cmpRepartition(getInstance(hist),instance))
    test_coeur = (Ncoeur is None or int(getCoeurs(hist)) == Ncoeur)
    test_algo = (algo is None or getAlgo(hist) == algo)
    test_sat = (sat is None or getSatellites(hist)==sat)
    test_seed = (seed_instance is None or getSeedInstance(hist)==seed_instance)
    return (test_instance and test_coeur and test_algo and test_sat and test_seed and test_name)
                
def filtrer(samples,name=None,instance=None,Ncoeur=None,algo=None,bestCoeur=False,sat=None,seed_instance=None):
    res = []
    for hist in samples:
        # hist,name=None,instance=None,Ncoeur=None,algo=None,sat=None,seed_instance=None
        if filtreUnitaire(hist,name,instance,Ncoeur,algo,sat,seed_instance):
            res.append(hist)
    if bestCoeur:
        new = []
        # les instances
        inst = []
        algos = []
        for h in res:
            if getInstance(h) not in inst:
                inst.append(getInstance(h))
            if getAlgo(h) not in algos:
                algos.append(getAlgo(h))
        # run de duree min
        for h in res:
            pass
            #print(h.algo_name,h.instance)
        #print(inst,algos)
        cmp = lambda h1 : h1.historique[-1][0]
        for i in inst:
            for a in algos:
                #sam = [h for h in res if h.instance.split('.txt')[0]==i.split('.txt') and h.algo_name==a]
                sam = [h for h in res if getInstance(h)==i and getAlgo(h)==a]
                if(len(sam))>0:
                    new.append(min(sam,key=cmp))
                else:
                    print("liste vide")
        return new
    else:
        return res

def getSeedInstance(sample):
    t1 = sample['instance'].split('seed')[1]
    t2 = t1.split("-")[1]
    return int(t2.split("_")[0])

def couperHistorique(sample,duree):
    ss = deepcopy(sample)
    tLimReach = ss['hist'][-1][0]>duree
    if tLimReach:
        ss['hist'] = [h for h in ss['hist'] if h[0]<= duree]
        last = ss['hist'][-1]
        last = tuple([duree]+[last[i] for i in range(1,len(last))])
        ss['hist'].append(last)
    for cpu in ss['stats']['envoi']:
        ss['stats']['envoi'][cpu] = [x for x in ss['stats']['envoi'][cpu] if x[0]<=duree]
        ss['stats']['calculs'][cpu] = [x for x in ss['stats']['calculs'][cpu] if x[0]<=duree]
        ss['stats']['reception'][cpu] = [x for x in ss['stats']['reception'][cpu] if x[0]<=duree]
    return ss

def getRepartitionRequete(instance_name):
    sp = instance_name.split("_")[0].split("-")
    return tuple([int(x) for x in sp])
                                           
def printTextBf(valeur,extreme_val,end):
    if valeur==extreme_val:
        msg = "\\textbf{"+str(valeur)+"}"
    else:
        msg = str(valeur)
    print(msg,end=end)

def getTaille(sample):
    instance_name = sample['instance']
    sp = instance_name.split("_")
    return sum((int(sp[3]),int(sp[4]),int(sp[5]),int(sp[6])))
    
def tracerCPU(hist,contour=True,legend=True,xlim=None,firstit_only=True):
    f,ax = plt.subplots(figsize=(15,6))
    #plt.xlabel('time (s)')
    #plt.ylabel('CPU index')
    tps_max = 0
    tps_min = np.Inf
    
    limite = max([x[0] for x in hist['hist']])
    if firstit_only:
        limite = [x for x in hist['hist'] if x[1]=="END_ITERATION"][0][0]
    
    for cpu in hist['stats']['calculs']:
        for (t_start,duree) in hist['stats']['calculs'][cpu]:
            if t_start<=limite:
                tps_max = max(tps_max,t_start+duree)
                tps_min = min(tps_min,t_start)
                if contour:
                    calcul_handle = ax.add_patch(patches.Rectangle((t_start, cpu),duree,1,edgecolor='black',facecolor = 'green',fill=True,label='utilisation du solveur') )
                else:
                    calcul_handle = ax.add_patch(patches.Rectangle((t_start, cpu),duree,1,facecolor = 'green',fill=True,label='utilisation du solveur') )
    for cpu in hist['stats']['reception']:
        for (t_start,duree) in hist['stats']['reception'][cpu]:
            if t_start<=limite:
                tps_max = max(tps_max,t_start+duree)
                tps_min = min(tps_min,t_start)
                rcv_handle = ax.add_patch(patches.Rectangle((t_start, cpu),duree,1,facecolor = 'red',fill=True,label='message') ) # slave => master
    for cpu in hist['stats']['envoi']:
        for (t_start,duree) in hist['stats']['envoi'][cpu]:
            if t_start<=limite:
                tps_max = max(tps_max,t_start+duree)
                tps_min = min(tps_min,t_start)
                send_handle = ax.add_patch(patches.Rectangle((t_start, cpu),duree,1,facecolor = 'red',fill=True,label='message') )         # maitre => slave   
    if legend:
        plt.legend(handles=[send_handle,calcul_handle],loc='lower right') #rcv_handle
    t = 'cpu occupation : '+hist['instance']
    #plt.title(t)
    plt.yticks(range(int(hist['coeurs'])))
    plt.ylim(0,max(list(hist['stats']["envoi"].keys()))+1)
    if xlim is None:
        plt.xlim(tps_min,tps_max)
    else:
        plt.xlim(tps_min,xlim)
    #plt.savefig("../results/cpu/"+self.algo_name+"_"+self.instance+"_"+str(self.Ncoeur)+".png")
    return f

def getObjectif(hist,tlim=None):
    if tlim is None:
        if hist["algo"] == "timeDependentSolver":
            return hist['hist'][-1][2][0]
        else:
            return max([x[2][0] for x in hist['hist']])
    else:
        if hist["algo"] == "timeDependentSolver":
            return hist['hist'][-1][2][0]
        else:
            return max([x[2] for x in hist['hist'] if x[0]<=tlim+0.05])
```

```python
def compareParametres(sample_path,instances_tableau,tlim,Nsat,dossiers,noms,algo ="BPCCAS"):
    print("\\begin{table*}[!h]")
    print("\\centering")
    print("\\footnotesize")
    print("\\begin{tabular}{|p{1.95cm}|p{.7cm}|",end='')
    
    for nom in noms:
        print("p{.5cm}p{.5cm}p{.5cm}|",end='')
    print("}")
    samples = []
    for k,d in enumerate(dossiers):
        samples.append(getSamples(sample_path+d))
    all_instances = []
    for sample_d in samples:
        for sample in sample_d:
            if sample["instance"] not in all_instances:
                all_instances.append(sample["instance"])
    print("\\hline")
    print("instance & \\#A &",end='')
    for k,nom in enumerate(noms):
        print("\\multicolumn{3}{c|}{"+nom+"}",end='')
        if k!=len(noms)-1:
            print("&",end='')
        else:
            print("\\\\")
    print("& &",end='')
    for k,nom in enumerate(noms):
        print("rew. & rew. 1st & dur. 1st.",end='')
        if k!=len(noms)-1:
            print("&",end='')
        else:
            print("\\\\")
    print("\\hline")
    for l,instance in enumerate(instances_tableau):
        # Infos de l'instance
        print('\t'+str(instance[0])+"\_"+str(instance[1])+"\_"+str(instance[2])+"\_"+str(instance[3]),end=" & ")
        instance_graine = sorted([i for i in all_instances if getRepartitionRequete(i)==instance and getSatFromName(i)==Nsat])
        samples_cca = []
        for i in instance_graine:
            samples_cca += filtrer(samples[0],instance=i,sat=Nsat)
        try:
            print(int(np.mean([getNActivites(x) for x in samples_cca])),end=" & ")
        except:
            print("?",end=" & ")
        # rÃ©sultat de chaque dossier (contient une version avec des parametres)
        reward = {}
        duree = {}
        reward_first_it = {}
        prof_mean = {}
        prof_max = {}
        for k,sample_d in enumerate(samples):
            res_samples = []
            for j,i in enumerate(instance_graine):
                res_samples += [filtrer(sample_d,instance=i,algo=algo,sat=Nsat,Ncoeur=10)[0]]
            reward[noms[k]] = round(np.mean([getObjectif(x) for x in res_samples]),1)
            prof_mean[noms[k]] = round(np.mean([profondeurModeMax(x) for x in res_samples]),1)
            prof_max[noms[k]] = round(np.mean([profondeurModeMoyen(x) for x in res_samples]),1)
            reward_first_it[noms[k]] = round(np.mean([getFirstIteration(x)[1][0] for x in res_samples]),1)
            try:
                if len([getFirstIteration(x)[0] for x in res_samples])>0:
                    duree[noms[k]] = round(np.mean([getFirstIteration(x)[0] for x in res_samples]),1)
                else:
                    duree[noms[k]] = tlim
            except:
                duree[noms[k]] = tlim
                    #if l>=6:
                    #    duree[algo] =  300
                #print(duree[algo])
                #duree[algo] = int(duree[algo])
                #print(np.mean([getDateFin(x) for x in res_samples]))
        duree = {d:int(duree[d]) for d in duree}
        max_reward = max([reward[d] for d in reward])
        min_dur = min([duree[d] for d in duree])
        max_first = max ([reward_first_it[d] for d in reward_first_it])
        for k,sample_d in enumerate(samples):
            if k==len(samples)-1:
                term = ""
            else:
                term = " & "
            printTextBf(reward[noms[k]],max_reward,' & ')
            #print(prof_mean[nom[k]],end=' & ')
            #print(prof_max[nom[k]],end=term)
            printTextBf(reward_first_it[noms[k]],max_first,' & ')
            printTextBf(duree[noms[k]],min_dur,term)
            #if algo!='globalSolver':
                #printTextBf(reward_first_it[algo],np.Inf,' & ')
                #printTextBf(duree[algo],-np.Inf,term)


        print("  \\\\") 
    print("\\hline \n")
    print("\\end{tabular}")
    print("\\normalsize")
    print("\\end{table*}")
              
def getSamplesFolderList(root,folders,noms,tlim=None):
    assert(len(folders)==len(noms))
    samples = {}
    all_instances = []
    for i,folder in enumerate(folders):
        samples[noms[i]] = getSamples(root+folder)
    for nom in samples:
        for s in samples[nom]:
            if s['instance'] not in all_instances:
                all_instances.append(s['instance'])
            if tlim is not None and s['algo']!="globalSolver":
                s = reduire(s,tlim)
    return samples,all_instances

def afficherNombreActivites(instances,Nsat,samples_folder,seed_instance=None,end=" & "):
    samples = samples_folder[list(samples_folder.keys())[0]]
    samples_cca = []
    for i in instances:
        samples_cca += filtrer(samples,name=i,sat=Nsat,seed_instance=seed_instance)
    try:
        print(int(np.mean([getNActivites(x) for x in samples_cca])),end=end)
    except:
        print("?",end=end)
    """
    try:
        print(int(np.mean([getCCA(x) for x in samples_cca])),end=end)
    except:
        print("?",end=end)    
    """   
def getRequetesSatisfaites(sample):
    return len(sample["modes"])
    
def extraireResultatsAlgos(samples,noms,algos,instances_candidates,Nsat,tlim,seed_instance=None,gap=False):
    def extraireReward(reward,nom,res_samples,tlim):
        reward[nom] = round(np.mean([getObjectif(x,tlim) for x in res_samples]),1)
    def extraireProfModes(prof_mean,prof_max,nom,res_samples):
        prof_mean[nom] = round(np.mean([profondeurModeMoyen(x) for x in res_samples]),1)
        prof_max[nom] = round(np.mean([profondeurModeMax(x) for x in res_samples]),1) 
    def extraireCompteursAppels(op_calls,perturb_calls,nom,res_samples):
        res = [compteursIterations(sample) for sample in res_samples]
        try:
            op_calls[nom],perturb_calls[nom] = int(np.mean([x[0] for x in res])),int(np.mean([x[1] for x in res]))
        except ValueError:
            op_calls[nom],perturb_calls[nom] = 0,0
    def extraireActiviteRetirees(removed,nom,res_samples):
        try:
            removed[nom] = int(np.mean([sample['info_add']['activitÃ©s retirÃ©es'] for sample in res_samples]))
            
        except:
            removed[nom] = -1
    def extraireNombreRequetesSatisfaites(req_satisfaites,nom,res_samples):
        res = [getRequetesSatisfaites(sample) for sample in res_samples]
        try:
            req_satisfaites[nom] = int(np.mean(res))
        except ValueError:
            req_satisfaites[nom] = 0
    def getDureeFirstIt(sample):
        i = 0
        while i<len(sample['hist']):
            if sample['hist'][i][1]=="END_ITERATION":
                break
            else:
                i += 1
        if i<len(sample['hist']):
            return sample['hist'][i][0]
        else:
            return sample['hist'][-1][0]
    def getRewardFirstIt(sample):
        i = 0
        while i<len(sample['hist']):
            if sample['hist'][i][1]=="END_ITERATION":
                break
            else:
                i += 1
        if i<len(sample['hist']):
            return max([x[2] for x in sample['hist'][:i+1]])
        else:
            return sample['hist'][-1][2]
    def fillDur(duree,nom,res_samples):
        try:
            if len([getFirstIteration(x)[0] for x in res_samples])>0:
                duree[nom] = round(np.mean([getFirstIteration(x)[0] for x in res_samples]),1)
            else:
                duree[nom] = -1
        except:
            duree[nom] = tlim
    def getReducedSamples(samples,nom,algo,Nsat,instances_candidates,seed_instance=seed_instance):
        res_samples = []
        for x in instances_candidates:
            filtre_repartition_requetes = filtrer(samples[nom],name=x,instance=None,algo=algo,sat=Nsat,seed_instance=seed_instance)
            if tlim is not None:
                for i in range(len(filtre_repartition_requetes)):
                    filtre_repartition_requetes[i] = reduire(filtre_repartition_requetes[i],tlim)
            res_samples += filtre_repartition_requetes
        return res_samples
    def fillFirstIt(reward_first_it,dur_first_it,reward,duree):
        if algo!="globalSolver":
            reward_first_it[nom] = round(np.mean([getRewardFirstIt(x)[0] for x in res_samples]),1)
            dur_first_it[nom] = round(np.mean([getDureeFirstIt(x) for x in res_samples]),1)
        else:
            reward_first_it[nom] = reward[nom]
            dur_first_it[nom] = duree[nom]
    reward,duree,reward_first_it,dur_first_it,prof_mean,prof_max = {},{},{},{},{},{}
    op_calls = {}
    perturb_calls = {}
    removed = {}
    req_satisfaites = {}
    try:
        for nom,algo in zip(noms,algos):
            res_samples = getReducedSamples(samples,nom,algo,Nsat,instances_candidates,seed_instance=seed_instance)
            extraireReward(reward,nom,res_samples,tlim)
            extraireProfModes(prof_mean,prof_max,nom,res_samples)
            fillDur(duree,nom,res_samples)
            fillFirstIt(reward_first_it,dur_first_it,reward,duree)
            extraireCompteursAppels(op_calls,perturb_calls,nom,res_samples)
            extraireActiviteRetirees(removed,nom,res_samples)
            extraireNombreRequetesSatisfaites(req_satisfaites,nom,res_samples)
    except Exception as e:
        print(e)
        print(algo)
        print([profondeurModeMax(x) for x in res_samples])
        raise e
    duree = {nom:int(duree[nom]) for nom in duree}
    max_reward = max([reward[nom] for nom in reward])
    min_dur = min([duree[nom] for nom in duree])
    max_first = max ([reward_first_it[nom] for nom in reward_first_it])
    if gap:
        reward = {x:round(100*(max_reward-reward[x])/max_reward,2) for x in reward}
    return reward,duree,reward_first_it,prof_mean,prof_max,max_reward,min_dur,max_first,dur_first_it,op_calls,perturb_calls,removed,req_satisfaites

def extractConfig(sample,attribute,attribute_type):
    lines = sample['config'].split('\n')
    find = False
    for line in lines:
        split = line.split(":")
        if attribute in split[0]:
            find = True
            value = split[1]
            break
    value = value.split(" ")[1]
    if not find:
        for line in lines:
            split = line.split(":")
        raise NameError("Attribut introuvable")
        
    if(attribute_type!=str):
        value = eval(value)
    return attribute_type(value)

def bestAt(sample,tlim):
    return max([h for h in sample['hist'] if h[0]<=tlim],key = lambda h:h[2])

def printSamplesTableau(sample_path,instances_tableau,tlim,folders,noms,algos,shift_instances=False,seeds_instance=range(5),Nsat=16,ignore_systematic=True,gap=False,colonnes = ["instance","activites","reward"],info=False,cut=False): 
    def formatInstance(instance,seed,ignore_systematic,shift_instances):
        if shift_instances:
            start="&"
        else:
            start=""
        if ignore_systematic:
            somme = sum([instance[i] for i in range(5) if i!=3])
            return start+str(somme)+"&"+str(instance[0])+"&"+str(instance[1])+"&"+str(instance[2])+"&"+str(instance[4])
        else:
            somme = sum(instance)
            return start+str(somme)+"&"+str(instance[0])+"&"+str(instance[1])+"&"+str(instance[2])+"&"+str(instance[3])+"&"+str(instance[4])
    allowed = ["instance","activites","duration","mean mode","worst mode","ndr","removed","reward",'first it reward','first it dur','requetes satisfaites']
    for c in colonnes:
        if not(c in allowed):
            raise ValueError(str(c)+" n'est pas une colonne. Colonnes possibles : "+str(allowed))
    if algos is not None:
        assert(len(algos)==len(noms))
    samples,all_instances = getSamplesFolderList(sample_path,folders,noms,tlim)
    if info:
        print("=====================================")
        print("info du tableau")
        for i,nom in enumerate(samples):
            print(nom,"("+algos[i]+"):",len(samples[nom]),"instances")
        print("=====================================")
    for instance in instances_tableau:
        for seed in seeds_instance:
            print("\t",end='')
            if 'instance' in colonnes:
                #print('& ',end= '')
                print(formatInstance(instance,seed,ignore_systematic,shift_instances=shift_instances),end="")
            instances_candidates = sorted([i for i in all_instances if getRepartitionRequete(i)==instance and getSatFromName(i)==Nsat])
            if "activites" in colonnes:
                print(' & ',end= '')
                afficherNombreActivites(instances_candidates,Nsat,samples,seed_instance=seed,end="")
            reward,duree,reward_first_it,prof_mean,prof_max,max_reward,min_dur,max_first,dur_first_it,op_calls,perturb_calls,removed,req_satisfaites = extraireResultatsAlgos(samples,noms,algos,instances_candidates,Nsat,tlim,seed_instance=seed,gap=gap)
            if gap:
                print(" & ",max_reward,end='')
            for k,nom_algo in enumerate(noms):
                if "reward" in colonnes:
                    print(' & ',end='')
                    if gap:
                        if float(reward[nom_algo])==0.0:
                            print("\\textbf{"+str(0.0)+"}",end='')
                        else:
                            print(reward[nom_algo],end='')
                    else:
                        printTextBf(reward[nom_algo],max_reward,'')
                    if "removed" in colonnes and algos[k]=='timeDependentSolver':
                        print("&"+str(removed[nom_algo]),end="")
                    if 'ndr' in colonnes and algos[k]=='LNS':
                        print("&"+str(op_calls[nom_algo]),end="")
                    if 'requetes satisfaites' in colonnes:
                        print("("+str(req_satisfaites[nom_algo])+")",end="")
                if 'duration' in colonnes:
                    print(' & ',end='')
                    print(duree[nom_algo],end='')
                if 'first it reward' in colonnes:
                    if nom_algo in ['BPCCAS','UPCCAS']:
                        print(' & ',end='')
                        print(reward_first_it[nom_algo],end='')
                if 'first it dur' in colonnes:
                    if nom_algo in ['BPCCAS','UPCCAS']:
                        print(' & ',end='')
                        print(dur_first_it[nom_algo],end='')
                if "mean mode" in colonnes:
                    print(' & ',end='')
                    print(prof_mean[nom_algo],end='')
                if "worst mode" in colonnes:
                    print(' & ',end='')
                    print(prof_max[nom_algo],end='')

            print("  \\\\ \n",end="")   
```
